## Se importan las librerías a usar

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Se setean las funciones de pertenencia

```python
# Lineal Decreciente desde 1 en 0 hasta 0 en a1
def FPertenenciaLD(a1, y): 
    y = np.array(y) 
    pertenencia = ((a1 - y) / a1) * ((y >= 0) & (y <= a1))
    pertenencia[y > a1] = 0

    return pertenencia

# Lineal Decreciente desde 1 en a0 hasta 0 en a1.
def FPertenenciaLD2(a0, a1, x):
    x = np.array(x)
    pertenencia = np.zeros_like(x, dtype=float)
    pertenencia[x <= a0] = 1
    mask = (x > a0) & (x <= a1)
    pertenencia[mask] = (a1 - x[mask]) / (a1 - a0)
    return pertenencia

# Lineal Decreciente 
def FPertenenciaLD3(a1, y):
    y = np.array(y)
    pertenencia = np.zeros_like(y, dtype=float)
    mask = (y >= 0) & (y <= a1)
    pertenencia[mask] = (a1 - y[mask]) / a1
    return pertenencia

# Lineal Creciente desde 0 en a0 hasta 1 en a1.
def FPertenenciaLC(a0, a1, x):
    x = np.array(x)
    pertenencia = np.zeros_like(x, dtype=float)
    mask = (x > a0) & (x <= a1)
    pertenencia[mask] = (x[mask] - a0) / (a1 - a0)
    pertenencia[x > a1] = 1
    return pertenencia
```

## Se setean las funciones de graficado

```python
# Graficar función de pertenencia (Lineal Decreciente)
def plot_FPertenenciaLD(a, a1, b, y, title, label):
    plt.plot(y, FPertenenciaLD(a1, y), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a], [1], color='red', s=60)
    plt.scatter([a1, b], [0, 0], color='red', s=60)
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()

# Graficar función de pertenencia (Lineal Decreciente)
def plot_FPertenenciaLD2(a0, a1, x, title, label):
    plt.plot(x, FPertenenciaLD2(a0, a1, x), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a0], [1], color='red', s=60, marker='x')
    plt.scatter([a1, b], [0, 0], color='red', s=60, marker='x')
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()

# Graficar función de pertenencia (Lineal Decreciente)
def plot_FPertenenciaLD3(a, a1, b, y, title, label):
    plt.plot(y, FPertenenciaLD3(a1, y), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a], [1], color='red', s=60)
    plt.scatter([a1, b], [0, 0], color='red', s=60)
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()
    
# Graficar función de pertenencia (Lineal Creciente)
def plot_FPertenenciaLC(a, a0, a1, x, title, label):
    plt.plot(x, FPertenenciaLC(a0, a1, x), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a], [0], color='red', s=60)
    plt.scatter([a0], [0], color='red', s=60)
    plt.scatter([a1, b], [1, 1], color='red', s=60)
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()

# Graficar la función de pertenencia (Lineal Creciente)
def plot_FPertenenciaLC2(a0, a1, b, x, title, label):
    plt.plot(x, FPertenenciaLC(a0, a1, x), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a0], [0], color='red', s=60, marker='x')
    plt.scatter([a1, b], [1, 1], color='red', s=60, marker='x')
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()
```
