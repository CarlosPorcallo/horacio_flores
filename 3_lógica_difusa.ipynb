## Se importan las librerías a usar

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Se setean las funciones de pertenencia difusa

```python
# Lineal Decreciente
def FPertenenciaLD(a1, y):
    y = np.array(y) 
    pertenencia = ((a1 - y) / a1) * ((y >= 0) & (y <= a1))
    pertenencia[y > a1] = 0

    return pertenencia

# Lineal Decreciente
def FPertenenciaLD2(a0, a1, x):
    x = np.array(x)
    pertenencia = np.zeros_like(x, dtype=float)
    pertenencia[x <= a0] = 1
    mask = (x > a0) & (x <= a1)
    pertenencia[mask] = (a1 - x[mask]) / (a1 - a0)
    return pertenencia

# Lineal Decreciente
def FPertenenciaLD3(a1, y):
    y = np.array(y)
    pertenencia = np.zeros_like(y, dtype=float)
    mask = (y >= 0) & (y <= a1)
    pertenencia[mask] = (a1 - y[mask]) / a1
    return pertenencia

# Lineal Creciente
def FPertenenciaLC(a0, a1, x):
    x = np.array(x)
    pertenencia = np.zeros_like(x, dtype=float)
    mask = (x > a0) & (x <= a1)
    pertenencia[mask] = (x[mask] - a0) / (a1 - a0)
    pertenencia[x > a1] = 1
    return pertenencia
```

## Se setean las funciones de graficado

```python
# Graficar función de pertenencia (Lineal Decreciente)
def plot_FPertenenciaLD(a, a1, b, y, title, label):
    plt.plot(y, FPertenenciaLD(a1, y), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a], [1], color='red', s=60)
    plt.scatter([a1, b], [0, 0], color='red', s=60)
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()

# Graficar función de pertenencia (Lineal Decreciente)
def plot_FPertenenciaLD2(a0, a1, x, title, label):
    plt.plot(x, FPertenenciaLD2(a0, a1, x), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a0], [1], color='red', s=60, marker='x')
    plt.scatter([a1, b], [0, 0], color='red', s=60, marker='x')
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()

# Graficar función de pertenencia (Lineal Decreciente)
def plot_FPertenenciaLD3(a, a1, b, y, title, label):
    plt.plot(y, FPertenenciaLD3(a1, y), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a], [1], color='red', s=60)
    plt.scatter([a1, b], [0, 0], color='red', s=60)
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()
    
# Graficar función de pertenencia (Lineal Creciente)
def plot_FPertenenciaLC(a, a0, a1, x, title, label):
    plt.plot(x, FPertenenciaLC(a0, a1, x), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a], [0], color='red', s=60)
    plt.scatter([a0], [0], color='red', s=60)
    plt.scatter([a1, b], [1, 1], color='red', s=60)
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()

# Graficar la función
def plot_FPertenenciaLC2(a0, a1, b, x, title, label):
    plt.plot(x, FPertenenciaLC(a0, a1, x), label=r'$\mu_{A}(x)$', color='blue', linewidth=4)
    plt.scatter([a0], [0], color='red', s=60, marker='x')
    plt.scatter([a1, b], [1, 1], color='red', s=60, marker='x')
    plt.xlabel(label)
    plt.ylabel(r'$\mu_{A}(x)$')
    plt.grid(True)
    plt.title(title)
    plt.show()
```

## Conjuntos Difusos para el Conjunto de Datos "Sequía"

```python
# Definición de parámetros
a = 0 # Valor inicial para generar una secuencia de puntos
a1 = 40 # Umbral clave para definir la función de pertenencia (límite de cambio)
b = 60 # Valor Final de la secuencia de puntos
y = np.arange(a, b + 1, 1) # Secuencia de puntos enteros: 0 a 60 con paso de 1
title = "Función de pertenencia difusa: Sequía"
label = "Drought (Index)"

# Definición de Evaluaciones
valores = [
    [6, 14, 0, 17, 19, 5, 33, 12, 5, 7, 1, 5],
    [6, 19, 1, 18, 27, 6, 34, 15, 4, 0, 0, 3],
    [5, 20, 1, 14, 19, 5, 34, 11, 3, 2, 1, 3],
    [6, 17, 0, 19, 26, 6, 32, 14, 4, 3, 0, 3],
    [5, 20, 1, 17, 23, 10, 34, 13, 5, 2, 0, 3]
]

# Graficar función de pertenencia
plot_FPertenenciaLD(a, a1, b, y, title, label)

# Imprimir resultados de pertenencia
for i, datos in enumerate(valores, 1):
    print(f"- Evaluación {i}: {FPertenenciaLD(a1, datos)}")
    
```
